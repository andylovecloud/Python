with the **unique**-function it is easier to look for incorrect or suspicious values

Unique displays the unique values of the column as a list. For example, the following shows that the 'sat_colleques' column has values 2, 3, 4, and 5 and one or more missing values (nan). The values are printed in the order they are received in the data.

```python
df['sat_colleques'].unique()
```

```python
# With the for-loop we can go thought all the columns (except number now)
# by default the drop method removes rows (axis=0), but now we remove columns (axis=1)
for column in df.drop('number', axis=1):
    print(column, df[column].unique())
```

```python
# Three rows where the salary has the biggest values
df.nlargest(3, 'salary')
```

```python
#  Three rows where the salary has the smallest values
df.nsmallest(3, 'salary')
```

The **value_counts** function counts the frequencies - hence the number of occurences for uniq values

Now, let's calculate the frequencies for all the variables/columns but for number column.

```python
for var in df.drop('number', axis=1):
    print(df[var].value_counts())
```

```python
# The descriptive statistics for the data is also worth checking
df.describe()
```

## List of functions and functionalities used in this notebook

* **pd.read_excel()** Reading data into notebook/Python from the Excel file
* **df.shape** The dimensions/count of rows and columns
* **pd.options.display.max_rows = None** For showing all the rows
* **pd.options.display.max_columns = None** For showing all the columns
* **df.head()** head of data
* **df.tail()** tail of data
* **df.count()** the column names and counts for values for each column
* **df.isnull().sum()** The amount of missing values for each column
* **df.info()** Names of the columns, counts and data types for columns
* **df['column_name'].unique()** The list of uniq values for the column_name
* **df.drop()** For removing columns or rows
* **print()** print to the screen
* **df.nlargest(n, column)** n largest values for the column
* **df.nsmallest(n, column)** n smallest values for the column
* **df['column_name'].value_counts()** The counting the frequencies for the column_name
* **df.describe()** Descriptive statistics for DataFrame
