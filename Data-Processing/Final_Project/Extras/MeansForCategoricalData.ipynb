Next, we calculate mean for categorical variable - more precisely for a variable measured in ordinal scale!
But we have discussed that then the values are not numerical and we should not calculate for example the mean.
Why do we do that then???

Actually, this is a thing that there is a lot of depate whether this is allowed or not. The other party says no and
justifies that by saying that the level of measurement is only ordinal and we can not be sure whether the distance between various options is always
the same (is option 0 as far from option 1 as is option 1 from option 2, for example). And the others are saying
that when measuring opinions, the scale for measurement can however considered as somewhat evenly distributed and hence
it would be ok, to use mean here.

Practice has however shown that we get somewhat good results with surveys measuring opinions and with calculating means for them.

Ok, this is maybe then up to you to choose, how do you wish :)

But let's see what we could do if we admit a bit of flexibility for us and start calculating the mean (or maybe some other measures).
